```python
import pandas as pd
df = pd.read_csv('fused_gemm.csv')
df["ThreadBlockShape"] = [(m, n, k) for m, n, k in zip(df["ThreadBlockShapeM"], df["ThreadBlockShapeN"], df["ThreadBlockShapeK"])]
df["WarpShape"] = [(m, n, k) for m, n, k in zip(df["WarpShapeM"], df["WarpShapeN"], df["WarpShapeK"])]
df.drop(columns=["ThreadBlockShapeM", "ThreadBlockShapeN", "ThreadBlockShapeK", "WarpShapeM", "WarpShapeN", "WarpShapeK"], inplace=True)
df_fused = df[df['Fused']]
df_fused["Time"] = df_fused["ShmemFusionTime"]
df_fused.drop(columns=["ShmemNonFusionTime", "ShmemFusionTime"], inplace=True)
df_non_fused = df[df['Fused']==False]
df_non_fused["Time"] = df_non_fused["ShmemNonFusionTime"]
df_non_fused.drop(columns=["ShmemNonFusionTime", "ShmemFusionTime"], inplace=True)
df_fused.sort_values(by='Time', inplace=True)
df_non_fused.sort_values(by='Time', inplace=True)
```
